(1) For the LTI systems described by the following difference equations, generate its impulse response, and 
unit step response. Comment on the properties of the system. 
\begin{equation}\nonumber
(a)~~y[n]-0.3695y[n-1]+0.1958y[n-2]=0.2066x[n]+0.4131x[n-1]+0.2066x[n-2]
\end{equation}
\begin{equation}\nonumber
 (b)~~~~~~~y[n]=0.77y[n-1]+x[n]+x[n-1]
\end{equation}
\begin{equation}\nonumber
 (c)~y[n]=0.77y[n-1]+0.77x[n]+x[n-1] 
\end{equation}

```python
%matplotlib inline
```

```python
(a)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from scipy.io.wavfile import write
from scipy.io import wavfile
system=signal.dlti([0.2066, 0.4131, 0.2066],[1.0, -0.3695, 0.1958],)
t,h=signal.dimpulse(system, n=50)
plt.figure(figsize=(15,5))
plt.stem(t,np.squeeze(h),use_line_collection=True)
plt.title('Impulse Response', color='b')
plt.xlabel('n'+ r'$\rightarrow$')
plt.ylabel('h[n]'+ r'$\rightarrow$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xticks(np.arange(-5,51))
plt.show()
t,s=signal.dstep(system, n=50)
plt.figure(figsize=(15,5))
plt.stem(t,np.squeeze(s),use_line_collection=True)
plt.title('Step Response', color='b')
plt.xlabel('n'+ r'$\rightarrow$')
plt.ylabel('s[n]'+ r'$\rightarrow$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xticks(np.arange(-5,51))
plt.show()
```

(3) Find the convolution of the following  sequences. Plot the input and output sequences

\begin{equation}\nonumber
(a)~~~~~~~~~~~~~~~h[n]=u[n]-u[n-8],x[n]=u[n]-u[n-8]
\end{equation}
\begin{equation}\nonumber
(b)~~~~~~~~~h[n]=u[n]-u[n-8],x[n]=2^n (u[n]-u[n-8])
\end{equation}
\begin{equation}\nonumber
(c)~~h[n]=u[n]-u[n-8],x[n]=\cos(\frac{\pi n}{4})(u[n]-u[n-8]) 
\end{equation}

Use \textit{conv} funtion to find convolution.

```python
(a)
```

```python
#Input signals x[n]=h[n]=u[n]-u[n-8]
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt 
u=np.ones(20)
u_8=np.ones(20)
u_8[0:8]=0
h=u-u_8
x=u-u_8
y=np.convolve(h,x)
plt.figure(figsize=(20,5))
plt.stem(x)
plt.title('x[n] = h[n]', color='b')
plt.xlabel('n'+ r'$\rightarrow$')
plt.ylabel('x[n] = h[n] '+ r'$\rightarrow$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xticks(np.arange(-10,20))
plt.show()
#The plot below are the input signals which are both equal
```

```python
plt.figure(figsize=(20,5))
plt.stem(y)
plt.title('y[n]', color='b')
plt.xlabel('n'+ r'$\rightarrow$')
plt.ylabel('y[n]'+ r'$\rightarrow$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xticks(np.arange(-10,20))
plt.show()
#The plot below is the convolution of the 2 signals x[n] and h[n].
```

```python
#Input signal h[n]=u[n]-u[n-8]
#Input signal x[n]=2^n(u[n]-u[n-8])
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
n=np.arange(20)
u=np.ones(20)
u_8=np.ones(20)
u_8[0:8]=0
h=u-u_8
x=np.power(2,n)*h
y=np.convolve(h,x)
plt.figure(figsize=(20,5))
plt.stem(x)
plt.title('x[n]', color='b')
plt.xlabel('n'+ r'$\rightarrow$')
plt.ylabel('x[n]'+ r'$\rightarrow$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xticks(np.arange(-10,20))
plt.figure(figsize=(20,5))
plt.stem(h)
plt.title('h[n]', color='b')
plt.xlabel('n'+ r'$\rightarrow$')
plt.ylabel('h[n] '+ r'$\rightarrow$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xticks(np.arange(-10,20))
plt.show()
#The plots below are those of the 2 input signals x[n] and h[n]
```

```python
plt.figure(figsize=(20,5))
plt.stem(y)
plt.title('y[n]', color='b')
plt.xlabel('n'+ r'$\rightarrow$')
plt.ylabel('y[n] '+ r'$\rightarrow$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xticks(np.arange(-10,20))
plt.show()
#The plot below is the convolution of the 2 signals x[n] and h[n]
```

```python
#Input signal h[n]=u[n]-u[n-8]
#Input signal x[n]=cos(n*pi/4)(u[n]-u[n-8])
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
n=np.arange(20)
u=np.ones(20)
u_8=np.ones(20)
u_8[0:8]=0
h=u-u_8
x=np.cos((n*np.pi)/4)*h
y=np.convolve(h,x)
plt.figure(figsize=(20,5))
plt.stem(x)
plt.title('x[n]', color='b')
plt.xlabel('n'+ r'$\rightarrow$')
plt.ylabel('x[n]'+ r'$\rightarrow$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xticks(np.arange(-10,20))
plt.figure(figsize=(20,5))
plt.stem(h)
plt.title('h[n]', color='b')
plt.xlabel('n'+ r'$\rightarrow$')
plt.ylabel('h[n] '+ r'$\rightarrow$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xticks(np.arange(-10,20))
plt.show()
#The plots below are those of the 2 input signals x[n] and h[n]
```

```python
plt.figure(figsize=(15,5))
plt.stem(y)
plt.title('y[n]', color='b')
plt.xlabel('n'+ r'$\rightarrow$')
plt.ylabel('y[n] '+ r'$\rightarrow$')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xticks(np.arange(-10,16))
plt.show()
#The plot below is the convolution of the 2 signals x[n] and h[n]
```
